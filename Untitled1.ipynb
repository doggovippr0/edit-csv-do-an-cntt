### Thêm thư viện Pandas

```python
import pandas as pd 
```

### Tạo hàm tách field a bởi dấu ',' và đảo ngược, sau đó đưa tới index thứ x

```python
def fieldshift(df,a,x,b,y):
    df[['first','last']] = df[a].str.split(b,expand=True)
    df.fillna('',inplace=True)
    if (y == 1):
        df['full'] = df['last'] + ' ' + df['first']
    elif (y == 0):
        df['full'] = df['first'] + ' ' + df['last']
    df = df.drop(['first','last', a], 1)
    second_column = df.pop('full')
    df.insert(x, a, second_column)
    return df
```

### Đọc file và áp dụng hàm, xuất file csv output

```python
df = pd.read_csv('books_new_N2.csv',delimiter=',')
display(df)
df = fieldshift(df,'Author',1,',',1)
df = fieldshift(df,'Title',0,',',1)
df = fieldshift(df,'SubGenre',3,'_',0)
df.to_csv('dulieuchuanhoa.csv')
```

### Tách file csv đã xử lí thành 1.csv và 2.csv

```python
csvfile = open('dulieuchuanhoa.csv', 'r').readlines()
#mo access
#mo table, ko co tao moi
filename = 1
for i in range(len(csvfile)):
    if i % 110 == 0:
        open(str(filename) + '.csv', 'w+').writelines(csvfile[i:i+110])
        filename += 1
```
