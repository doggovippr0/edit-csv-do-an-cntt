```python
import pandas as pd 
```

```python
df = pd.read_csv("books_new_N2.csv",delimiter=',')
```

```python
df[['Last name','First name']] = df['Author'].str.split(',',expand=True)
```

```python
df.fillna("",inplace=True)
```

```python
df['Full name of Author'] = df['First name'] + ' ' + df['Last name']
```

```python
df = df.drop(['Author', 'First name', 'Last name'], 1)
```

```python
second_column = df.pop('Full name of Author')
df.insert(1, 'Author', second_column)
```

```python
df[['Book name','A/The']] = df['Title'].str.split(',',expand=True)
```

```python
df.fillna("",inplace=True)
```

```python
df['Full name of book'] = df['A/The'] + ' ' + df['Book name']
```

```python
df = df.drop(['Book name', 'A/The','Title'], 1)
```

```python
second_column = df.pop('Full name of book')
df.insert(0, 'Title', second_column)
```

```python
df.to_csv('dulieuchuanhoa.csv')
```
